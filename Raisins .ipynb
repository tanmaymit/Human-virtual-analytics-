## Altair Implementation of Prototype 1 

```python
import pandas as pd
import altair as alt
raisins = pd.read_excel('Raisin_Dataset.xlsx')
```

```python
raisins.head()
```

```python
click = alt.selection_multi(encodings=['color'])

line = alt.Chart(raisins).mark_line().encode(
    x='Area:Q',
    y='MajorAxisLength:Q',
    color='Class:N',
    tooltip='MajorAxisLength'
).transform_filter(
    click
).interactive()


scatter = alt.Chart(raisins).mark_point().encode(
    x='Area:Q',
    y='Extent:Q',
    color='Class:N',
    tooltip='Extent'
).transform_filter(
    click
).interactive()


hist = alt.Chart(raisins).mark_bar().encode(
    x='count()',
    y='Class',
    color=alt.condition(click, 'Class', alt.value('lightgray'))
).add_selection(
    click
)

scatter & line & hist
```
